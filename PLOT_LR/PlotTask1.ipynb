## Задание 1

**Условие:**
- Постройте график функции $y=2*\sin x + 4$.
- Подпишите оси и заголовок графика

**Решение:**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 24, 500)

y = 2 * np.sin(x) + 4

plt.plot(x, y)
plt.show()
````

## Задание 2

**Условие:**
- Постройте графики функций $y=x^2$ и $y=x$ на одном графике.
* Дайте название графику
* Дайте названия осям
* Присвойте лейблы и выведите легенду
* Отрисуйте графики функций пунктирными линиями и разными цветами


**Решение:**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 500)

y1 = x
y2 = x ** 2

plt.figure()

plt.title("График прямой и параболы")

plt.plot(x, y1, color='royalblue', linestyle='--', label="Прямая")
plt.plot(x, y2, 'r--', label="Парабола")

plt.xlabel("Абсцисса")
plt.ylabel("Ордината")

plt.legend()

plt.show()
````

## Задание 3

**Условие:**
Создайте случайную выборку из ```1000000``` значений, полученную в результате реализации случайной величины, распределенной по нормальному закону с параметрами ```[0, 1]```. Постройте гистограмму для полученного массива с параметром ```bins=100```. Укажите среднее значение в полученном массиве.

**Решение:**

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(0, 1, size=1000000)

print("Среднее значение", np.median(x))

plt.hist(x, bins=100)
plt.show()


````

## Задание 4

**Условие:**
- Создайте массив ```DataFrame```, используя код ниже.
- Создайте ```relplot```  с параметром ```hue='c'```, взяв два оставшихся столбца в качестве значений по осям координат.

**Решение:**

````python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.DataFrame()
df['a'] = np.random.rand(100)
df['b'] = np.random.rand(100)
df['c'] = np.random.randint(0,5,100)
df.head()

sns.relplot(df, x='a', y='b', hue='c')
plt.show()
````

## Задание 5

**Условие:**
Для датафрейма из задания 4:
- Постройте ```countplot``` для столбца ```'c'```.
- Постройте ```displot``` для значений каждого столбца. Не перепутайте с ```distplot```.

**Решение:**

````python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.DataFrame()
df['a'] = np.random.rand(100)
df['b'] = np.random.rand(100)
df['c'] = np.random.randint(0,5,100)
df.head()

plt.figure()

sns.countplot(df, x='c')
plt.show()

sns.displot(df['a'])
plt.show()

sns.displot(df['b'])
plt.show()

sns.displot(df['c'])
plt.show()
````
