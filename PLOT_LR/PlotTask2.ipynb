## Задание 1

**Условие:**
Постройте графики функций $y=e^x$ и $y=x^e$ на одной координатной сетке. Проделайте следующие операции:

* дайте названия графикам
* дайте названия осям
* присвойте лейблы и выведите легенду
* отрисуйте графики функций пунктирными линиями и разными цветами


**Решение:**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 500)

y1 = np.exp(x)
y2 = x ** np.e

plt.figure()

plt.title("Графики e ** x и x ** e")

plt.plot(x, y1, 'r--', label="e ** x")
plt.plot(x, y2, 'b--', label="x ** e")

plt.xlabel("Абсцисса")
plt.ylabel("Ордината")

plt.legend()

plt.show()
```

## Задание 4
**Условие:**
Постройте гистограмму для первого признака в датасете выше следующим образом:
* на одном графике должно быть три гистограммы, по одной на каждый класс,
* для этого используйте методологии подвыборки массива с условием,
* количество бинов должно определяться автоматически (режим "auto"),
* гистограмма должна быть читаемой, т. е. добавьте подписи, лейблы, названия и т. д.

### **Решение:**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
data = datasets.load_iris(return_X_y=False)
X = data.data
y = data.target
names = data.feature_names

plt.title("Гистограмма распределения первого признака по классам")
plt.xlabel("Значение первого признака")
plt.ylabel("Частота (количество значений)")

plt.hist(X[y == 0, 0], alpha=0.6, label='Класс 1', bins='auto')
plt.hist(X[y == 1, 1], alpha=0.6, label='Класс 2', bins='auto')
plt.hist(X[y == 2, 2], alpha=0.6, label='Класс 3', bins='auto')

plt.legend()

plt.show()
```

## Задание 5
**Условие:**
Посчитайте матрицу корреляции между признаками и нарисуйте ее как heatmap. Подберите подходящую по смыслу цветовую гамму. Добавьте названия признаков в качестве тиков по осям. Добавьте отрисовку цветовой шкалы.

### **Решение:**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd

data = datasets.load_iris(return_X_y=False)
X = data.data
names = data.feature_names

df = pd.DataFrame(X, columns=names)

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))

sns.heatmap(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

plt.show()
```
