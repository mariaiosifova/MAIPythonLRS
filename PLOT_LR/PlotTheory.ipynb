## Библиотека Matplotlib

  Начнем с библиотеки Matplotlib. Познакомимся с базовыми объектами данной библиотеки, с тем как сохраняются графики и как вообще конфигурируется объекте.
  Чаще всего на самом старте при анализе данных машинного обучения проходится строить самые примитивные графики каких либо функций или статистических распределений. Модуль Pyplot позволяет очень быстро создавать такие графики. Приступим!

###  Базовые объекты. Сохранение графиков. Конфигурирование Matplotlib

Посмотрим на пример построения простейшего графика:

```python
# Подгружаем модуль pyplot и библиотеку numpy
import matplotlib.pyplot as plt
import numpy as np

# Создадим массив значений от 0 до 9 в переменной data
data = np.arange(10)

# Вызываем функцию plot из модуля pyplot и строим график зависимости элемента от его индекса
plt.plot(data)
plt.show()
```

```python
# Теперь построим похожий график, но явно укажем зависимость передав два массива функции plot
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Используем для отрисовки графика функцию show()
plt.show()
```

Посмотрим как мы можем изменить стиль нашего графика, передав в функцию plot различные параметры для цвета, маркера и стиля линий.


```python
# Добавим описание графика
print('Поточеченый график в виде красных точек:')

# Создадим поточеченый график в виде красных точек
# Передадим дополнительный параметром 'ro' - поточечный график с красными точками
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')

# С помощью функции axis() мы можем задать максимальные и минимальные значения для осей
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
print('Различные типы и цвета отрисовки линий:')

# Зададим массив значений, с которым мы будем работать
t = np.arange(0., 5., 0.2)

# Создадим объект рисунок, чтобы отрисовать все графики в одной координатной плоскости
plt.figure()

# В качестве третьего параметра передаем различные значения для задания цвета графика и ее стиля.
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')

# Так же можно рисовать ступеньками, но нужен параметр drawstyle
plt.plot(t, t**3.2, c='y', drawstyle='steps')

# Дополнительно с помощью параметра 'marker' отрисуем каждую точку плюсом.
plt.plot(t, t**3.4, c='darkred', ls='--', marker='+')

plt.show()
```

```python
print('Толщина линий:')

# Также мы можем менять толщину линии с помощью параметра linewidth
plt.plot(t, t, linewidth=2)
plt.plot(t, t**2, linewidth=4)
plt.plot(t, t**3, linewidth=8)
plt.show()
```

Отлично, мы уже кое-что умеем. Графики получаются красивые, но не очень понятные, не хватает подписей. Давйте посмотрим как это можно исправить.

```python
plt.figure()

# С помощью функции title добавим общую подпись сверху на графике. Чаще всего здесь описывают саму зависимость.
plt.title('Различные показательные функции')

# Далее строим уже знакомые нам графики, но добавляя параметр label указываем как каждой из них подписать в легенде в верхнем левом углу по умолчанию
plt.plot(t, t, 'r--',label='linear')
plt.plot(t, t**2, 'bs',label='x^2')
plt.plot(t, t**3, 'g^',label='x^3')

# С помощью xlabel и ylabel задаем подписи к соответсвующим осям
plt.xlabel('Абсцисса')
plt.ylabel('Ордината')

# Используя legend выводим в верхний левый угол табличку с легендой
plt.legend()

plt.show()
```

Отлично, наши графики стали более осмысленными. А что делать, если мы хотим отобразить сразу несколько графиков рядом для наглядности? Давайте разберем несколько подходов реализации такой идеи.

```python
plt.figure()

# Для реализации нашей идеи используем функцию subplot. В качестве параметра передаем три цифры:
# 2 - количество строк, 1 - количество столбцов, 1 - какой по счету  график
plt.subplot(211)
plt.plot(t, t, 'g--')

# 2 - количество строк, 1 - количество столбцов, 2 - какой по счету график
plt.subplot(212)
plt.plot(t, t**4, 'rs')


plt.show()

# Данный способ не всегда работает стабильно, взглянем на еще один
```

```python
# Альтернативный метод

# С помощью метода subplots создаем рисунок и полотна. В качестве параметров передаем количество строк и количество столбцов
fig, axes = plt.subplots(2, 1)

# Выбираем по индексу 0 первое полотно и строим на нем сразу два графика
axes[0].plot(t, t, 'g--')
axes[0].plot(t, t ** 2, 'r--')

# Выбираем второе полотно по индексу 1 и строим один график
axes[1].plot(t, t**4, 'b')

plt.show()
```

```python
# Хотим больше графиков? Легко! Передаем  в subplots  количество строк 2 и количество столбцов 2
fig, axes = plt.subplots(2, 2)

# Теперь необходимо обращаться к полотнам по двум индексам
# Нарисуем четыре линейных графика
axes[0,0].plot(t, t, 'g--')
axes[0,1].plot(t, t, 'r--')
axes[1,0].plot(t, t, 'b--')
axes[1,1].plot(t, t, 'y--')
plt.savefig('ThreeGraphs.png')
plt.show()
```

Как строить различные графики мы с вами разобрались, отлично. Теперь научимся их сохранять. Для этого нам понадобится функция savefig.

```python
# Строим график
fig, axes = plt.subplots(2, 2)

axes[0,0].plot(t, t ** 2, 'g--')
axes[0,1].plot(t, t ** 2, 'r--')
axes[1,0].plot(t, t ** 2, 'b--')
axes[1,1].plot(t, t ** 2, 'y--')

# И сохраняем его с помощью функции savefig
plt.savefig('ThreeGraphs.png')

plt.show()
```

```python
# С помощью метода imread подгрузим сохраненный график и с помощью imshow отобразим его
image = plt.imread('ThreeGraphs.png')
plt.imshow(image)
```

### Примеры использования библиотеки Matplotlib

С базовыми конструкциями разобрались, теперь посмотрим на различные примеры использования библиотеки Matplotlib. Чаще всего вам понадобятся следующие виды графиков из данный библиотеки: диаграммы, гистограммы, тепловые карты и точечные диаграммы.

#### Диаграмма

Диаграмма — графическое представление данных линейными отрезками или геометрическими фигурами, позволяющее быстро оценить соотношение нескольких величин. Представляет собой геометрическое символьное изображение информации с применением различных приёмов техники визуализации.

Например, если у нас есть объем продаж некоторого товара по месяцам, то для анализа таких данных удобно будет построить диаграмму. По оси Х у нас будут отложены месяца, по оси Y соответствующие им объемы продаж.

```python
# Создадим массив для оси Х, где значение массива будет определять индекс определенного месяца
x = np.arange(5)

# Зададим объем продаж в каждом месяце
height = [20, 23, 25, 24, 26]

# Для построенеи диаграммы воспользуемся функцией bar. В качестве параметров передадим массивы значений X и Y, а также зададим цвет диаграммы
plt.bar(x, height, color='c')

# Подписываем график уже знакомыми нам функциями
plt.title('Продажи товара Х по месяцам')
plt.ylabel('млн руб.')
plt.xlabel('Месяц')

# Для подписи каждого столбца используем функцию xticks. Передаем сам массив X и кортеж подписей
plt.xticks(x, ('Январь', 'Февраль', 'Март', 'Апрель', 'Май'))

plt.show()
```

Дополнительные [примеры](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar) кода.

#### Гистограмма

Гистограмма -  способ представления статистических данных в графическом виде – в виде столбчатой диаграммы. Она отображает распределение отдельных измерений параметров изделия или процесса. Иногда ее называют частотным распределением, так как гистограмма показывает частоту появления измеренных значений параметров объекта.

Высота каждого столбца указывает на частоту появления значений параметров в выбранном диапазоне, а количество столбцов – на число выбранных диапазонов.

```python
# Для демонстрации воспользуемся нормальным распределением.
# х - массив нормально распределенной величины длиной 1000, с параметрами нормального распределения 0 и 1
x = np.random.normal(0, 1, 1000)

# Тогда как часто встречается каждое из значений в массиве х нам может показать гистограмма
# Строим гистограмму с помощью функции hist, передаем в качестве параметра массив х и указываем количество столбцов bins=20
plt.hist(x, bins=20)
plt.title('Гистограмма распределения значений в массиве х')
plt.show()
```

Дополнительные [примеры](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) кода.

#### Тепловая карта

Тепловые карты - это весьма эффективный и наглядный способ визуализации данных. Значение параметра показывается не количественно - с помощью цифр, а качественно - с помощью цвета. Иначе говоря, конкретные числовые значения преобразуются в цвета. И эти числовые значения обязательно привязаны к какой-либо точке или зоне.



```python
# Продемонстрируем работу Heatmap
# Cоздаем массив из нулей
my_array = np.zeros((9,9))

# добавляем строку единиц
my_array[4] = 1

# добавляем столбец единиц
my_array[:,4] = 1

# Посмотрим на полученный массив
print(my_array)
```

```python
# Выводим массив по цветам палитры c помощью метода imshow
plt.imshow(my_array)

#Добавим цветовую полосу справа от heatmap. Она нам покажет какой цвет имеет элемент в массиве в зависимости от значения
plt.colorbar()

plt.show()
```

```python
# Еще один пример heatmap

# Зададим двумерный массив размером 10х10 состоящий из случайных чисел от 0 до 1
arr = np.random.rand(10,10)

# Отрисуем в этот раз heatmap со стилем YlGn
plt.imshow(arr, cmap='YlGn')
plt.colorbar()
plt.show()
```

Рассмотрим как тепловая карта может помочь при анализе данных. Допустим у нас есть данные, сколько каждый фермер вырастил овощей каждого вида. Тогда очень удобно будет представить такую информацию в виде heatmap.

```python
print('Пример тепловой визуализации сбора урожая различных овощей фермерами:')

# Создаеем массивы подписей для овощей и фермеров
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

# Зададим массив значений объемов выращенных овощей
harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


# Создаем рисунок и полотно
fig, ax = plt.subplots(figsize=(7,7))

# Отрисовываем heatmap
im = ax.imshow(harvest)

# Отрисовываем штрихи
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))

# Наносим подписи к штрихам
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Поворачиваем вертикальные подписи для читабельности
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Подписываем каждую ячейку соответствующим значением
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

# Подписываем заголовок
ax.set_title("Harvest of local farmers (in tons/year)")

# Отрисовываем цветовую полосу справа от heatmap
ax.figure.colorbar(im, ax=ax)

# Функция tight_layout используется для лучшей визуализации, она предотвращает пересечение подписей
fig.tight_layout()
plt.show()
```

Из такого графика можем увидеть, например:
- Больше всего картофеля производит компания BioGoods Ltd.
- Компания Cornylee Corp. специализируется на ячмене(barley) и пшенице(wheat)
- Спаржа(asparagus) неинтересна фермерам

Дополнительные [примеры](https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html) кода.

#### Точечная диаграмма

Точечная диаграмма— математическая диаграмма, изображающая значения двух переменных в виде точек на декартовой плоскости. Диаграммы рассеяния используются для демонстрации наличия или отсутствия зависимости между двумя переменными.

Да, обычные графики тоже демонстрируют нам зависимость, но давайте посмотрим, в чем может быть преимущество точечной диаграммы.

```python
# Создадим два массива случайных величин размером N
N = 200
x = np.random.rand(N)
y = np.random.rand(N)

# Пусть z массив результатов произведений соответствующих элементов из x и y
z = x * y

# Зададим массив логических переменных по условию z > 0.2
# Далее это понадобится нам для разделения точек по цветам
colors = z > 0.2

# Создаем рисунок и полотно
fig, ax = plt.subplots(2,1, figsize=(6,6))

# Отрисовываем точечную диаграмму со следующими параметрами:
# x, y - массивы значений для соответсвующих осей
# s - размер точки, который считается отдельно для каждого значения из z
# c - задаем цвет, разбиваем точки на два класса с помощью массива логических значений colors
ax[0].scatter(x, y, s=100*z, c=colors)

# Устанливаем подпись к точечной диаграмме
ax[0].set_title('Scatter plot')

# Для сравнения отрисуем такой же график с помощью ранее изученной функции plot, параметр 'yo' - отрисовываем кружками желтого цвета
ax[1].plot(x, y, 'yo')
ax[1].set_title('Basic plot with yellow circles')

fig.tight_layout()
plt.show()
```

Вот в чем преимущества точечной диаграммы. Во-первых, она умеет отображать трехмерные графики на двумерной плоскости, размерами точек она может отобразить дополнительное условие. Во-вторых, цветом может отобразить принадлежность определенному классу. Например, в данном случае мы классифицировали точки по параметру z. Попробуйте поменять условие для z и посмотреть как изменится цвет точек.

## Библиотека Seaborn

Seaborn — библиотека для создания статистических графиков на Python. Она построена на основе matplotlib и тесно интегрируется со структурами данных pandas. Seaborn помогает вам изучить и понять данные. Его функции построения графиков работают с датасетами и выполняют все необходимы преобразования для создания информативных графиков.

Синтаксис, ориентированный на набор данных, позволяет сосредоточиться на графиках, а не деталях их построения. Библиотека Seaborn предоставляет набор датасетов, на которых можно обучаться работать с самой библиотекой и ставить эксперименты по анализу данных. Список доступных для экспериментов датасетов можно найти по этой [ссылке](https://github.com/mwaskom/seaborn-data).


### Базовые графики

Подгрузим библиотеку Seaborn и  базу чаевых. Построим на ней различные виды графиков для анализа этих данных

```python
# Импортирум саму библиотеку Seaborn и указываем, что будем использовать ее под псевдонимом sns
import seaborn as sns

# Задаем общий стиль графиков
sns.set(style="ticks")

# Подргужаем базу чаевых, которая уже есть как пример в библиотеке Seaborn
tips = sns.load_dataset("tips")

# Выводим с помощью метода head первые пять записей
tips.head()

#Данные, которые мы получим в таблице:
# total_bill - общая стоимость заказа, tip - размер чаевых, sex - пол, smoker - курящий человек или нет,
# day - день недели, time - время приема пищи, size - кол-во позиций в заказе
```

```python
# С помощью функции replot построим график зависимости чаевых от стоимости заказа
# Передаем в значения параметров:  x - стоимость заказа, y - размер чаевых,
# hue - укажем разбиение на цветовые классы в соответствии с днем недели, data - указываем с каким набором данных мы работаем
g = sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
```

По графику выше заметим, что чем больше стоимость заказа, тем больше чаевых можно получить. Можно предположить, что имеет место линейная зависимость. Далее в разделе графики регрессии попробуем в этом убедиться.

Допустим, что теперь вы хотим оценить сколько чаевых можно получить за ужин и обед отдельно. Для используем параметр col:

```python
# Строим точно такой же график, но теперь с помощью параметра col="time" разделим по значениям столбца time на два столбца, в данном случае есть на обед и ужин.
g = sns.relplot(x="total_bill", y="tip", hue="day",
                col="time", data=tips)
```

По анализу данного графика заметим, например, что на обед ходят чаще всего по четвергам, а на ужин по субботами и воскресеньям, а также, что размер чаевых больше на ужин. Добавим разделение по полу, для этого воспользуемся параметров row:

```python
# Добавим к нашему графику параметр row="sex", в котором укажем способ разбиение по строкам. Получим некоторую матрицу для четырех возможных сочетаний
g = sns.relplot(x="total_bill", y="tip", hue="day",
                col="time", row="sex", data=tips)
```

### Визуализация отношений категориальных данных

Отлично, теперь разберемся как делать визулизацию некоторых категориальных данных. Для этого снова воспользуемся обученным датасетом из билиотеки Seaborn, а именно данными о пассажирах Титаника.

#### "Ящик с усами" или Boxplot



Boxplot – это удобный способ визуализации распределения данных по
квартилям. Эти усы указывают на изменчивость за пределами верхнего и нижнего

Общая схема boxplot представлена ниже:
- график boxplot показывает средний, нижний и верхний квартили,
- минимальное и максимальное значение выборки и выбросы

Построим пример такого boxplot'а для зависимости возраста от пола и с разбиением на классы пассажиров:

```python
# Для этого воспользуемся функцией boxplot, куда в качесте y передадим пол, в качестве x возраст, а в качестве разбиения пола укажем в hue класс,
# а также укажем горизонтальную ориентацию orient="h"
sns.boxplot(x="age", y="sex", hue="class", orient="h", data=titanic)
```

### Графики регрессии

Ранее мы анализировали датасет с данными о чаевых. Мы преположили, что имеет место линейная зависимость чаевых от суммы заказа, с помощью lmplot попробуем в этом убедиться. Данная функция пытается по предоставленным данным восстановит линейную регрессию.

```python
# Передадим функции lmplot данные: x - сумма заказа, y - чаевые, data - датафрейм из которого мы берем указанные столбцы
sns.lmplot(x="total_bill", y="tip", data=tips)
```

Такой график несет не очень много информации, попробуй сделать разделение по одному из признаков. Например, курит клиент или нет. Для этого дополнительно воспользуемся параметром hue:

```python
# Передаем в параметр hue значение из столбца smoker для такого разделения
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

Отлично, теперь мы можем увидеть, что больше чаевых оставляют люди, которые не курят. Исходя из этой информации можем запустить антитабачную рекламу, некоторые клиенты бросят курить и будут платить нам больше чаевых :)

Покажем как мы можем отрисовать графики на отдельных полотнах:

```python
# Для этого воспользуемся известным нам параметром col, который разобьет данные на два полотна в зависимости от признака smoker
sns.lmplot(x="total_bill", y="tip", col="smoker", data=tips)
```

### Тепловые карты для визуализации матричных данных

Опять же идея использования тепловых карт точно такая же как и в случае библиотеки Matplotlib, но в Seaborn имеет преимущество именно в удобстве обработки матричных данных.

Посмотрим как построить тепловую карту на случаных значениях в Seaborn и воспользуемся тепловой картой в примере с Титаником:

```python
# Создадим матрицу 10х12 и заполним ее сулчайными числами от 0 до 1
uniform_data = np.random.rand(10, 12)

# С помощью функции heatmap построим тепловую карту на полученных данных
ax = sns.heatmap(uniform_data)
```

Для того, чтобы применить тепловую карту к данным о Титанике, сначала вытащим из таблицы интересующие нас данные. С помощью метода pivot_table создадим таблицу количества выживших в зависимости от пола и класса пассажира:

```python
# Вызываем функци pivot_table куда передаем столбец с меткой выживших, пол и класс. В aggfunc указывваем как нам поступить с данными о выживших, мы их суммируем для каждого сочетания
tit_srv = titanic.pivot_table('survived', 'sex', 'class', aggfunc=np.sum)

# Выводим полученную таблицу
tit_srv.head()
```

Теперь отрисуем это в виде тепловой карты:

```python
sns.heatmap(tit_srv)
```

Из это таблицы мы можем сделать вывод, что больше всего выжило женщин из первого класса, а меньше всего мужчин из второго. Таким образом можно использовать тепловую карту в анализе данных.

### Визуализация попарных отношений признаков в данных

Очень удобно отрисовать сразу все возможные интерисующие нас зависимости, для это мы можем использовать функцию pairplot. Воспользуемся ей для уже известных нам примеров:

```python
# Передаем в функцию pairplot датасет чаевых. Получаем построение попарных зависимостей
sns.pairplot(tips)
```

Если по оси X и по оси Y указана одна и та же величина, то строится гистограмма для значений данной величины. Такой прием используется, чтобы получить визуализацию сразу большого числа зависимостей, а далее можно выбрать конкретную и отдельно с ней работать.

Теперь воспользуемся данным методом для датасета пассажиров Титаника. Выберем с помощью параметра vars столбцы, для которых мы будем строить попарные зависимости, а с помощью параметра hue мы можем указать, если нужно, способ разделения полученных данных.

```python
# В параметр vars столбцы датафрейма, от которых мы хотим получить попарные зависимости, а в hue способ разделения данных, в данном случае это класс пассажира
sns.pairplot(titanic, hue='pclass', vars=['age', 'fare', 'survived'])
```
